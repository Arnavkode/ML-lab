```python
#question 1

import numpy as np

#(A)
arr = np.array([1,2,3,4,5,6])
rev = arr[::-1]
print(rev)

#(b)
array1 = np.array([[1,2,3], [2,4,5], [1,2,3]])

print(array1.flatten())
print(array1.ravel())

#(C)
arr1 = np.array([[1,2], [3,4]])
arr2 = np.array([[1, 2], [3, 4]])
print(arr1 > arr2)

#(d)
x = np.array([1,2,3,4,5,1,2,1,1,1]) 
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ]) 


unique_x, counts_x = np.unique(x, return_counts=True)
mode_x = unique_x[np.argmax(counts_x)]
indices_x = np.where(x == mode_x)[0]
print(f"For array x: The most frequent value is {mode_x}, found at indices {indices_x}")

unique_y, counts_y = np.unique(y, return_counts=True)
mode_y = unique_y[np.argmax(counts_y)]
indices_y = np.where(y == mode_y)[0]
print(f"For array y: The most frequent value is {mode_y}, found at indices {indices_y}")


#(E)
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(gfg.sum())
print(sum(gfg))
print(gfg.sum(axis = 1))

#(F)
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
print("Diagonal sum:", np.trace(n_array))
print("Eigen values: ", np.linalg.eigvals(n_array) )
print("Eigen vectors: ", np.linalg.eig(n_array) )
print("Inverse matrix: ", np.linalg.inv(n_array) )
print("Determinant: ", np.linalg.det(n_array) )

#(G)
p = [[1, 2], [2, 3]]
q = [[4, 5], [6, 7]] 

multiplication_1 = np.array(p) @ np.array(q)
print(f"Matrix multiplication (p1 @ q1):\n{multiplication_1}")
print("Covariance: ", np.cov(np.array(p), np.array(q)))

p1 = np.array([[1, 2], [2, 3], [4, 5]])
q1 = np.array([[4, 5, 1], [6, 7, 2]])

print(f"Matrix multiplication (p1 @ q1):", p1 @ q1)
# print("Covariance: ", np.cov(p1, q1)) ##GETTING PROBLEMS HERE


#(H)
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

print("Inner product: ", np.inner(x,y))
print("Outer product: ", np.outer(x,y))
print("Cartesian product: ", np.array(np.meshgrid(x.ravel(), y.ravel())).T.reshape(-1,2))


```

```python
## Question 2

#(A)
array = np.array([[1, -2, 3],[-4, 5, -6]])
print(np.abs(array))
print("Percentiles for flattened arrays: ", np.percentile(array.flatten(), [25,50,75]))
print("Percentiles for column arrays: ", np.percentile(array, [25,50,75], axis = 0))
print("Percentiles for row arrays: ", np.percentile(array, [25,50,75], axis = 1))

print("Mean of flattened : ", np.mean(array.flatten(), axis = 0), "Median of flattened: ", np.median(array.flatten(), axis= 0), "Standard deviation of flattened: ", np.std(array.flatten(), axis = 0))
print("Mean of each column: ", np.mean(array, axis = 0), "Median of each column: ", np.median(array, axis= 0), "Standard deviation of each column: ", np.std(array, axis = 0))
print("Mean of each row: ", np.mean(array, axis = 1), "Median of each row: ", np.median(array, axis= 1), "Standard deviation of each row: ", np.std(array, axis =1))

#(b)
a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])
print(np.floor(a))
print(np.ceil(a))
print(np.round(a))

```

```python
#Question 4
from PIL import Image as pillow

count = 0

def img_to_array(path):

    img = pillow.open(path)
    img_array = np.array(img)
    count = count + 1

    if img.mode == "L":
        print("Image in Grayscale")
        np.savetxt('grayscale_image_{count}.txt', img_array, fmt='%d')

    elif img.mode == "RGB":
        print("Image in RGB mode")
        reshaped_array = img_array.reshape(-1, 3)
        np.savetxt("RGB_image_{count}.txt", reshaped_array, fmt='%d')


def load_imagearray(path):
    image_array = np.loadtxt(path)
    print(image_array)
```
