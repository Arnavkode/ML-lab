```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the Iris dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target

# Split the dataset into training and testing sets (80:20 ratio)
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

print("Iris dataset loaded and split successfully.")
print(f"X_train_iris shape: {X_train_iris.shape}")
print(f"X_test_iris shape: {X_test_iris.shape}")
print(f"y_train_iris shape: {y_train_iris.shape}")
print(f"y_test_iris shape: {y_test_iris.shape}")
```

```python
from sklearn.svm import SVC

# Create an instance of SVC with a linear kernel
linear_svc_iris = SVC(kernel='linear', random_state=42)

# Train the model on the Iris training data
linear_svc_iris.fit(X_train_iris, y_train_iris)

print("Linear SVM model trained successfully on Iris dataset.")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions on the test set
y_pred_iris_linear = linear_svc_iris.predict(X_test_iris)

# Calculate accuracy
accuracy_linear = accuracy_score(y_test_iris, y_pred_iris_linear)

# Generate classification report
report_linear = classification_report(y_test_iris, y_pred_iris_linear)

# Generate confusion matrix
conf_matrix_linear = confusion_matrix(y_test_iris, y_pred_iris_linear)

print(f"Linear SVM Model Performance on Iris Dataset:\n")
print(f"Accuracy: {accuracy_linear:.4f}\n")
print("Classification Report:\n")
print(report_linear)
print("Confusion Matrix:\n")
print(conf_matrix_linear)
```

```python
from sklearn.svm import SVC

# Create an instance of SVC with a polynomial kernel
poly_svc_iris = SVC(kernel='poly', degree=3, random_state=42) # Default degree is 3

# Train the model on the Iris training data
poly_svc_iris.fit(X_train_iris, y_train_iris)

print("Polynomial SVM model trained successfully on Iris dataset.")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions on the test set using the polynomial SVM
y_pred_iris_poly = poly_svc_iris.predict(X_test_iris)

# Calculate accuracy
accuracy_poly = accuracy_score(y_test_iris, y_pred_iris_poly)

# Generate classification report
report_poly = classification_report(y_test_iris, y_pred_iris_poly)

# Generate confusion matrix
conf_matrix_poly = confusion_matrix(y_test_iris, y_pred_iris_poly)

print(f"Polynomial SVM Model Performance on Iris Dataset:\n")
print(f"Accuracy: {accuracy_poly:.4f}\n")
print("Classification Report:\n")
print(report_poly)
print("Confusion Matrix:\n")
print(conf_matrix_poly)
```

```python
from sklearn.svm import SVC

# Create an instance of SVC with an RBF kernel
rbf_svc_iris = SVC(kernel='rbf', random_state=42)

# Train the model on the Iris training data
rbf_svc_iris.fit(X_train_iris, y_train_iris)

print("RBF SVM model trained successfully on Iris dataset.")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions on the test set using the RBF SVM
y_pred_iris_rbf = rbf_svc_iris.predict(X_test_iris)

# Calculate accuracy
accuracy_rbf = accuracy_score(y_test_iris, y_pred_iris_rbf)

# Generate classification report
report_rbf = classification_report(y_test_iris, y_pred_iris_rbf)

# Generate confusion matrix
conf_matrix_rbf = confusion_matrix(y_test_iris, y_pred_iris_rbf)

print(f"RBF SVM Model Performance on Iris Dataset:\n")
print(f"Accuracy: {accuracy_rbf:.4f}\n")
print("Classification Report:\n")
print(report_rbf)
print("Confusion Matrix:\n")
print(conf_matrix_rbf)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load the Breast Cancer dataset
breast_cancer = load_breast_cancer()
X_cancer = breast_cancer.data
y_cancer = breast_cancer.target

# Split the dataset into training and testing sets (80:20 ratio)
X_train_cancer, X_test_cancer, y_train_cancer, y_test_cancer = train_test_split(X_cancer, y_cancer, test_size=0.2, random_state=42)

print("Breast Cancer dataset loaded and split successfully.")
print(f"X_train_cancer shape: {X_train_cancer.shape}")
print(f"X_test_cancer shape: {X_test_cancer.shape}")
print(f"y_train_cancer shape: {y_train_cancer.shape}")
print(f"y_test_cancer shape: {y_test_cancer.shape}")
```

```python
from sklearn.svm import SVC

# Create an instance of SVC with an RBF kernel
rbf_svc_cancer_no_scale = SVC(kernel='rbf', random_state=42)

# Train the model on the Breast Cancer training data without scaling
rbf_svc_cancer_no_scale.fit(X_train_cancer, y_train_cancer)

print("RBF SVM model trained successfully on Breast Cancer dataset without scaling.")
```

```python
from sklearn.metrics import accuracy_score

# Make predictions on the training set
y_train_pred_cancer_no_scale = rbf_svc_cancer_no_scale.predict(X_train_cancer)

# Make predictions on the test set
y_test_pred_cancer_no_scale = rbf_svc_cancer_no_scale.predict(X_test_cancer)

# Calculate training accuracy
training_accuracy_no_scale = accuracy_score(y_train_cancer, y_train_pred_cancer_no_scale)

# Calculate testing accuracy
testing_accuracy_no_scale = accuracy_score(y_test_cancer, y_test_pred_cancer_no_scale)

print(f"RBF SVM (No Scaling) Training Accuracy: {training_accuracy_no_scale:.4f}")
print(f"RBF SVM (No Scaling) Testing Accuracy: {testing_accuracy_no_scale:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Instantiate StandardScaler
scaler = StandardScaler()

# Fit the scaler to the training data and transform both training and testing data
X_train_cancer_scaled = scaler.fit_transform(X_train_cancer)
X_test_cancer_scaled = scaler.transform(X_test_cancer)

# Create an instance of SVC with an RBF kernel
rbf_svc_cancer_scaled = SVC(kernel='rbf', random_state=42)

# Train the model on the scaled training data
rbf_svc_cancer_scaled.fit(X_train_cancer_scaled, y_train_cancer)

print("RBF SVM model trained successfully on Breast Cancer dataset with scaling.")
print(f"X_train_cancer_scaled shape: {X_train_cancer_scaled.shape}")
print(f"X_test_cancer_scaled shape: {X_test_cancer_scaled.shape}")
```

```python
from sklearn.metrics import accuracy_score

# Make predictions on the scaled training set
y_train_pred_cancer_scaled = rbf_svc_cancer_scaled.predict(X_train_cancer_scaled)

# Make predictions on the scaled test set
y_test_pred_cancer_scaled = rbf_svc_cancer_scaled.predict(X_test_cancer_scaled)

# Calculate training accuracy
training_accuracy_scaled = accuracy_score(y_train_cancer, y_train_pred_cancer_scaled)

# Calculate testing accuracy
testing_accuracy_scaled = accuracy_score(y_test_cancer, y_test_pred_cancer_scaled)

print(f"RBF SVM (Scaled) Training Accuracy: {training_accuracy_scaled:.4f}")
print(f"RBF SVM (Scaled) Testing Accuracy: {testing_accuracy_scaled:.4f}")
```
